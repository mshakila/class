# A very brief overview of Pandas for time series data

pandas has proven very successful as a tool for working with time series data, especially in the financial data analysis space. pandas provides a relatively compact and self-contained set of tools for performing common tasks related with time series data.

## Creation of date ranges in Pandas: the DatetimeIndex object

A time series is a measurement of one or more variables over a period of time and at a specific interval. Once a time series is captured, analysis is often performed to identify patterns in the data, in essence, determining what is happening as time goes by.

```python
# 60 periods from 2010 - 1 - 1
pd.date_range(datetime.date(2011,1,1), periods=60, freq='m') 
```

```python
pd.date_range('1990-1-1', periods=40, freq='A') 
```

### Creating a series with dates as index

```python
daily_range = pd.date_range(start=datetime.date(2014,1,1), end=datetime.date.today() , freq='d') 
```

```python
my_series = pd.Series(np.random.normal(size=len(daily_range)), index=daily_range)
```

```python
my_series.plot()
```

## Some time series functionality in Pandas

```python
yahoo = pd.read_csv('..\\datasets\\yahoo.csv')
```

```python
yahoo.head()
```

### Index by time:

```python
yahoo['Date'] = yahoo['Date'].apply(pd.to_datetime)
yahoo.set_index('Date', inplace=True)
```

```python
yahoo.head()
```

```python
yahoo.index
```

```python
yahoo.drop('Adj Close', axis = 1, inplace=True)
```

## Calculating new columns

```python
yahoo['day_range'] = yahoo['High'] - yahoo['Low']
yahoo['intraday_return'] = (yahoo['Close'] - yahoo['Open'])/yahoo['Open']
```

```python
yahoo.head()
```

## Quickly visualize

```python
yahoo['Close'].plot(title = 'Yahoo Stock price')
```

```python
yahoo['Volume'].plot()
```

```python
yahoo['intraday_return'].plot()
```

```python
yahoo['intraday_return'].plot(kind='hist', bins=20)
```

## Filter by values

```python
# find the days where the volume was more than 30M shares
yahoo[yahoo['Volume']>30e6]
```

### Is there a relashinshio between volume and range?

```python
yahoo.plot.scatter('Volume','day_range')
```

```python
yahoo.plot.scatter('Volume','intraday_return')
```

## Shifting a time series

```python
yahoo['Close'].shift(-5).head()
```

## resample: Grouping by time periods

```python
yahoo.resample('1M').mean()
```

### What was the volatility per month?

```python
yahoo.resample('1M').std()['Close'].plot(kind='bar')
```

```python
month_names = yahoo.resample('1M').std()['Close'].index.to_series().apply(lambda x: x.strftime('%B'))
```

```python
volatility = yahoo.resample('1M').std()['Close'][1:].plot(kind='bar', title='Volatility of Yahoo Stock 2016')
volatility.set_xticklabels(month_names[1:]);
```
